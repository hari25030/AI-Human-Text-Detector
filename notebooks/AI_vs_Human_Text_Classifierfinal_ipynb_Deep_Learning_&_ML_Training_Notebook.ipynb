```
# AI vs Human Text Detector Project Notebook
# This notebook is where I train all my AI models

# tools we need
import pandas as pd # for handling data tables
import numpy as np # for numbers stuff
import os # for making folders

# Tools for cleaning up text
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re # for finding patterns in text

# Tools for making words into numbers for the older ML models
from sklearn.feature_extraction.text import TfidfVectorizer

# My Machine Learning
from sklearn.svm import SVC # Support Vector Machine
from sklearn.tree import DecisionTreeClassifier # Decision Tree
from sklearn.ensemble import AdaBoostClassifier # AdaBoost

# Tools for testing
from sklearn.model_selection import train_test_split, GridSearchCV # split data, find best settings
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc # check how good models are

# Tools for drawing charts
import matplotlib.pyplot as plt
import seaborn as sns

# Tools for saving my models
import joblib # for saving ML models and the word-to-number tool
import pickle # for saving the tokenizer for DL models

# DEEP LEARNING STUFF
# PyTorch
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Keras helps with turning text into sequences (for DL)
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Create necessary directories
# makes sure the folders exist before we try to save files into them
os.makedirs('models', exist_ok=True) # Create 'models' folder if it doesn't exist
os.makedirs('data/training_data', exist_ok=True) # Create 'data/training_data' if it doesn't exist
print("Checked/Created 'models/' and 'data/training_data/' folders.")


#  Get the Text Data Ready
try:
    df = pd.read_excel('data/training_data/AI_vs_huam_train_dataset.xlsx')
    print("Yay! Data loaded successfully!")
except FileNotFoundError:
    print("Uh oh! Make sure 'AI_vs_huam_train_dataset.xlsx' is in 'data/training_data/'")
    print("Creating dummy data for now so we can continue...")
    # Dummy data for demonstration if file not found
    data = {'essay': [ # Changed 'text' to 'essay' here for dummy data consistency
        "This is an example of human-written text, with some unique phrasing and common grammatical errors.",
        "The quick brown fox jumps over the lazy dog.",
        "Artificial intelligence systems are designed to process large datasets and identify patterns efficiently.",
        "Neural networks simulate the human brain's structure to learn and make predictions.",
        "I love coding, it's so much fun and really challenging sometimes!",
        "The algorithm analyzed the textual content and generated a summary.",
        "My cat purred softly on my lap while I was reading a book.",
        "Deep learning models excel at tasks such as image recognition and natural language processing."
    ],
            'label': [0, 0, 1, 1, 0, 1, 0, 1]} # Changed to numbers for dummy data consistency
    df = pd.DataFrame(data)

# Show a bit of the data
print("A peek at my data:")
print(df.head())
print("\nHow many human vs AI texts I have:")
print(df['label'].value_counts())

# Make labels into numbers (0 for Human, 1 for AI)
df['label_num'] = pd.to_numeric(df['label'].astype(str).str.strip(), errors='coerce')

# Drop rows where label_num might have became NaN
print(f"\nBefore dropping NaNs in label_num: {len(df)} rows")
df.dropna(subset=['label_num'], inplace=True)
# Ensure label_num is integer type after dropping NaNs
df['label_num'] = df['label_num'].astype(int)
print(f"After dropping NaNs in label_num: {len(df)} rows")
print("Checking for NaNs in label_num again (should be 0):")
print(df['label_num'].isnull().sum())

# Verify that we still have data after dropping NaNs
if len(df) == 0:
    raise ValueError("DataFrame is empty after processing labels. Check your 'label' column in the Excel file for unexpected values.")


# Clean Up the Text
nltk.download('stopwords') # get common words to remove
nltk.download('wordnet') # get base forms of words
nltk.download('omw-1.4') # helper for wordnet

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower() # make all letters small
    text = re.sub(r'[^a-z\s]', '', text) # remove numbers and symbols
    words = text.split() # split text into words
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words] # remove common words and get base form
    return " ".join(words)

print("\nCleaning up text...")
df['clean_text'] = df['essay'].apply(clean_text) # Apply cleaning to the 'essay' column
print("Cleaned text examples:")
print(df['clean_text'].head())


# Prepare Data for Machine Learning Models (TF-IDF Vectorizer) ---
# TF-IDF turns words into numbers based on how important they are in the text
# and in all texts.
print("\nMaking words into numbers for ML models (TF-IDF)...")
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # pick top 5000 important words
X_tfidf = tfidf_vectorizer.fit_transform(df['clean_text']) # learn and transform
y_ml = df['label_num']

# Save this vectorizer, my app will need it!
joblib.dump(tfidf_vectorizer, 'models/vectorizer.joblib')
print("TF-IDF vectorizer saved!")


#  Prepare Data for Deep Learning Models (Tokenization and Padding)
print("\nMaking words into sequences for DL models...")
MAX_WORDS = 10000 # Max words to consider
MAX_LEN = 250 # Max length of a sentence

# Tokenizer learns all unique words and gives them numbers
tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<unk>") # <unk> for unknown words
tokenizer.fit_on_texts(df['clean_text']) # Fit tokenizer on clean_text, which comes from 'essay'

# Turn text into sequences of numbers
X_sequences = tokenizer.texts_to_sequences(df['clean_text'])

# Make all sequences the same length by adding zeros or cutting them
X_padded = pad_sequences(X_sequences, maxlen=MAX_LEN, padding='post', truncating='post')
y_dl = torch.tensor(df['label_num'].values, dtype=torch.float32) # labels for PyTorch

# Save the tokenizer
# Use pickle for tokenizer, as joblib might have issues with Keras Tokenizer
with open('models/tokenizer.pkl', 'wb') as f:
    pickle.dump(tokenizer, f)
print("Tokenizer saved!")


# Split Data for Training and Testing
# For ML models:
X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(X_tfidf, y_ml, test_size=0.2, random_state=42)

# For DL models:
X_train_dl, X_test_dl, y_train_dl, y_test_dl = train_test_split(X_padded, y_dl, test_size=0.2, random_state=42)


# Train and Tune Machine Learning models

# Helper to train and evaluate ML models
def train_and_evaluate_ml_model(model, param_grid, X_train, X_test, y_train, y_test, model_name):
    print(f"\n--- Training {model_name} ---")
    grid_search = GridSearchCV(model, param_grid, cv=3, n_jobs=-1, verbose=1, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    y_proba = best_model.predict_proba(X_test)[:, 1] # Probability for positive class (AI)

    print(f"Best settings for {model_name}: {grid_search.best_params_}")
    print(f"{model_name} Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"{model_name} Report:\n{classification_report(y_test, y_pred)}")

    # Confusion Matrix (how many right/wrong guesses)
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Human', 'AI'], yticklabels=['Human', 'AI'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('Actual Guess')
    plt.xlabel('My Guess')
    plt.show()

    # ROC Curve (how well it tells classes apart)
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.show()

    return best_model, roc_auc


# SVM Brain
svm_params = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
best_svm_model, svm_auc = train_and_evaluate_ml_model(SVC(probability=True), svm_params, X_train_ml, X_test_ml, y_train_ml, y_test_ml, "SVM")
joblib.dump(best_svm_model, 'models/svm_model.joblib')
print("SVM brain saved!")

#  Decision Tree Brain
tree_params = {'max_depth': [5, 10, 20, None], 'min_samples_leaf': [1, 5, 10]}
best_tree_model, tree_auc = train_and_evaluate_ml_model(DecisionTreeClassifier(), tree_params, X_train_ml, X_test_ml, y_train_ml, y_test_ml, "Decision Tree")
joblib.dump(best_tree_model, 'models/tree_model.joblib')
print("Decision Tree brain saved!")

# AdaBoost Brain
ada_params = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1]}
best_ada_model, ada_auc = train_and_evaluate_ml_model(AdaBoostClassifier(), ada_params, X_train_ml, X_test_ml, y_train_ml, y_test_ml, "AdaBoost")
joblib.dump(best_ada_model, 'models/ada_model.joblib')
print("AdaBoost brain saved!")


#  Train and Tune Deep Learning Brains (PyTorch)

# Helper for PyTorch datasets
class TextDataset(Dataset):
    def __init__(self, sequences, labels):
        self.sequences = sequences
        self.labels = labels

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        return torch.tensor(self.sequences[idx], dtype=torch.long), self.labels[idx]

# Create datasets and data loaders
train_dataset_dl = TextDataset(X_train_dl, y_train_dl)
test_dataset_dl = TextDataset(X_test_dl, y_test_dl)

BATCH_SIZE = 32
train_loader_dl = DataLoader(train_dataset_dl, batch_size=BATCH_SIZE, shuffle=True)
test_loader_dl = DataLoader(test_dataset_dl, batch_size=BATCH_SIZE, shuffle=False)

# Check if we have a GPU (fast computer part)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"\nUsing device: {device}")

# Common settings for DL models
VOCAB_SIZE = MAX_WORDS # how many unique words
NUM_CLASSES_DL_OUTPUT = 1 # Output for binary classification with BCEWithLogitsLoss
EMBEDDING_DIM = 128 # size of word vectors (my chosen size)
NUM_CLASSES = 2 # This is for overall classification (Human/AI), not model output


# CNN Brain
class CNNClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, num_classes_output, max_len):
        super(CNNClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0) # turn word numbers into vectors
        self.conv1 = nn.Conv1d(in_channels=embedding_dim, out_channels=128, kernel_size=5) # first filter
        self.relu = nn.ReLU() # activation function
        self.pool1 = nn.MaxPool1d(kernel_size=max_len - 5 + 1) # get most important feature from filter
        self.fc = nn.Linear(128, num_classes_output) # final layer for prediction (now 1 output)

    def forward(self, text):
        embedded = self.embedding(text).permute(0, 2, 1) # change shape for conv layer
        conved = self.relu(self.conv1(embedded))
        pooled = self.pool1(conved).squeeze(2) # remove extra dimension
        return self.fc(pooled) # final guess

#  Pass NUM_CLASSES_DL_OUTPUT to the model
cnn_model = CNNClassifier(VOCAB_SIZE, EMBEDDING_DIM, NUM_CLASSES_DL_OUTPUT, MAX_LEN).to(device)


#  LSTM Brain
class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_classes_output):
        super(LSTMClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True) # LSTM layer
        self.fc = nn.Linear(hidden_dim, num_classes_output) # final layer for prediction (now 1 output)

    def forward(self, text):
        embedded = self.embedding(text)
        lstm_out, (hidden, cell) = self.lstm(embedded)
        return self.fc(hidden.squeeze(0))


lstm_model = LSTMClassifier(VOCAB_SIZE, EMBEDDING_DIM, hidden_dim=256, num_classes_output=NUM_CLASSES_DL_OUTPUT).to(device)


#  RNN Brain
class RNNClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_classes_output):
        super(RNNClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, batch_first=True) # RNN layer
        self.fc = nn.Linear(hidden_dim, num_classes_output) # final layer for prediction (now 1 output)

    def forward(self, text):
        embedded = self.embedding(text)
        rnn_out, hidden = self.rnn(embedded)

        return self.fc(hidden.squeeze(0))


rnn_model = RNNClassifier(VOCAB_SIZE, EMBEDDING_DIM, hidden_dim=256, num_classes_output=NUM_CLASSES_DL_OUTPUT).to(device)


# Helper to train and evaluate DL models
def train_and_evaluate_dl_model(model, train_loader, test_loader, epochs, model_name):
    print(f"\n--- Training {model_name} ---")
    optimizer = optim.Adam(model.parameters()) # how the model learns
    criterion = nn.BCEWithLogitsLoss() # how we measure errors (for binary classification)

    best_accuracy = 0
    roc_auc = 0

    for epoch in range(epochs):
        model.train() # set model to training mode
        total_loss = 0
        for texts, labels in train_loader:
            texts, labels = texts.to(device), labels.to(device)
            optimizer.zero_grad() # clear old gradients
            predictions = model(texts).squeeze(1) # get guesses (now [batch_size] shape)
            loss = criterion(predictions, labels) # measure error
            loss.backward() # calculate gradients
            optimizer.step() # update model weights
            total_loss += loss.item()

        avg_loss = total_loss / len(train_loader)
        print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")

    # Evaluate the model
    model.eval() # set model to evaluation mode
    y_true_dl = []
    y_pred_dl = []
    y_proba_dl = []

    with torch.no_grad(): # don't calculate gradients during evaluation
        for texts, labels in test_loader:
            texts, labels = texts.to(device), labels.to(device)
            predictions = model(texts).squeeze(1) # Ensure predictions are [batch_size]
            # Convert logits to probabilities
            probabilities = torch.sigmoid(predictions)

            # Convert probabilities to binary predictions
            predicted_labels = (probabilities > 0.5).long()

            y_true_dl.extend(labels.cpu().numpy())
            y_pred_dl.extend(predicted_labels.cpu().numpy())
            y_proba_dl.extend(probabilities.cpu().numpy())

    accuracy = accuracy_score(y_true_dl, y_pred_dl)
    print(f"{model_name} Accuracy: {accuracy:.4f}")
    print(f"{model_name} Report:\n{classification_report(y_true_dl, y_pred_dl)}")

    # Confusion Matrix (how many right/wrong guesses)
    cm = confusion_matrix(y_true_dl, y_pred_dl)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Human', 'AI'], yticklabels=['Human', 'AI'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('Actual Guess')
    plt.xlabel('My Guess')
    plt.show()

    # ROC Curve (how well it tells classes apart)
    fpr, tpr, _ = roc_curve(y_true_dl, y_proba_dl)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.show()

    return model, roc_auc


#  Train CNN Brain
epochs = 5 # small number of epochs for quick demo
best_cnn_model, cnn_auc = train_and_evaluate_dl_model(cnn_model, train_loader_dl, test_loader_dl, epochs, "CNN")
torch.save(best_cnn_model.state_dict(), 'models/cnn_model.pth') # .pth for PyTorch models
print("CNN brain saved!")


#  Train LSTM Brain
best_lstm_model, lstm_auc = train_and_evaluate_dl_model(lstm_model, train_loader_dl, test_loader_dl, epochs, "LSTM")
torch.save(best_lstm_model.state_dict(), 'models/lstm_model.pth')
print("LSTM brain saved!")


#  Train RNN Brain
best_rnn_model, rnn_auc = train_and_evaluate_dl_model(rnn_model, train_loader_dl, test_loader_dl, epochs, "RNN")
torch.save(best_rnn_model.state_dict(), 'models/rnn_model.pth')
print("RNN brain saved!")

print("\n--- All Brains Trained and Saved! ---")
# download the .joblib, .pth, and .pkl files from the 'models' folder

```
